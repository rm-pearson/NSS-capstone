Narrow data down to juvenile fiction, because the data is *too big* for pandas!

```python
count = 0

for chunk in pd.read_csv('../Data/export.csv', dtype = {'col5': int, 'col7': str, 'col11': str}, chunksize = 10000):
    chunk_df = chunk
    chunk_df['Subjects'] = chunk_df['Subjects'].fillna('No subject information')
    chunk_df = chunk_df.loc[chunk_df['Subjects'].str.lower().str.contains(r'.*\bjuvenile.?n?o?n?.?fiction', regex = True)]

    if count == 0:
        chunk_df.to_csv('../Data.juvenile.csv', index = False)
        count = 1
    else:
        chunk_df.to_csv('../Data/juvenile.csv', index = False, mode = 'a', header = False)
```
